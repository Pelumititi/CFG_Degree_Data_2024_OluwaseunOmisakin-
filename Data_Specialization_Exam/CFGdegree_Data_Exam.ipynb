1. PANDAS (25 points)

```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe:
import pandas as pd

netflix_data = pd.read_csv('netflix_titles.csv')
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset
netflix_data.head(10)
```

```python
# 3 (1 mark). Show the last 5 rows

netflix_data.tail(5)
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.

num_complete_rows = netflix_data.notna().all(axis=1).sum()
num_complete_rows
```

```python
# Show the resulting Netflix DataFrame
complete_rows_df = netflix_data[netflix_data.notna().all(axis=1)]

complete_rows_df
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame. Use any method for this.

filtered_df = netflix_data[netflix_data['type'] == 'Movie']
filtered_df[['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description']]
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
  # duration of the movies and return the resulting data frame
netflix_data['duration'] = netflix_data['duration'].str.replace(' min', '')
# Extract number part from duration on data
netflix_data['duration'] = netflix_data['duration'].str.extract('(\d+)').astype(float)
```

```python
average_duration = netflix_data['duration'].mean()
average_duration
```

```python
netflix_data['duration'].fillna(average_duration, inplace=True)
netflix_data
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.

duplicate_rows = netflix_data[netflix_data.duplicated()]
duplicate_rows

# From the returned dataset below, there are no duplicate in the data
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.

netflix_data.columns = netflix_data.columns.str.title()
netflix_data.head()
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file.
netflix_data.to_csv('netflix_data_cleaned.csv', index=False)
```

2. MATPLOTLIB (25 points)

Use the netflix dataset to solve the exercises in this section.

```python
# 1 (10 marks). Create a visualisation (by using the plot that is most appropriate for this) which shows the frequency of movies by year of release.
import matplotlib.pyplot as plt

# Using the new saved cleaned data

new_netflix_data = pd.read_csv('netflix_data_cleaned.csv')

movies_per_year = new_netflix_data.groupby('Release_Year')['Title'].count().reset_index()
movies_per_year
```

```python
# Create the bar chart

plt.figure(figsize=(16, 6))
plt.bar(movies_per_year['Release_Year'], movies_per_year['Title'])
# Add labels and titles
plt.title('Movies Released By Year')
plt.xlabel('Released Year')
plt.ylabel('Number of Movies Released')

plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movis
  # in the dataset.
    
# Split the 'Listed_In' column by commas to get the series of genres
genres = new_netflix_data['Listed_In'].str.split(',').explode().str.strip()
# Count genre occurence
genre_counts = genres.value_counts()
# Define as in question
top_5_genres = genre_counts.head(5)

# Work towards the pie chart
plt.figure(figsize=(10, 7))
top_5_genres.plot.pie(autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(5)))
plt.title('Top 5 Most Popular Movie Genres')

plt.show()
```

```python
# 2 (5 marks). How many odd numbers do we have in the array? Show how we get this result using numpy
# Declare the odd number count
odd_no_count = 0

# Loop through the array
for num in arr:
    if num % 2 != 0:
        odd_no_count += 1

print(f"Odd number count: {odd_no_count}")
```
